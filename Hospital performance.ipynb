```python
#We are going to Analyse only for the Hospital in Ontario, So in the data "Type of hospital should not be blank"
df_ind = df_ind_org.loc[df_ind_org['Province/territory']=="Ontario"].dropna(axis=0, subset=['Type of hospital'])
df_ind.drop_duplicates()
```

```python
#Select the necessary fields and reindex
df_ind=df_ind.loc[df_ind['SES']=="All"][["Region","Type of hospital","Hospital or long-term care organization","Data year","Sex","Indicator","Indicator result","Unit of measurement"]]
df_ind
```

```python
#Check the indicators
df_ind['Indicator'].unique()
```

```python
#Select the indicators related to our Analysis
#1) Emergency Department Wait Time for Physician Initial Assessment (90% Spent Less, in Hours), Maasurement_units: Hour
#2) Total Time Spent in Emergency Department for Admitted Patients (90% Spent Less, in Hours), Maasurement_units: Hour
#3) In-Hospital Sepsis, Maasurement_units: per 1,000
#4) All Patients Readmitted to Hospital, Maasurement_units: Pecentage
#5) Hospital Deaths (HSMR), Maasurement_units: Numbers
df_ind=df_ind.sort_values(['Region','Type','Name','Year','Indicator','Sex'])
df_ind=df_ind.reset_index(drop=True)
df_ind
```

### Fill missing values with their median value

```python
median = df['Deaths'].median()
median = df['Deaths'].fillna(median, inplace=True)

median = df['Occupancy_Rate (%)'].median()
median = df['Occupancy_Rate (%)'].fillna(median, inplace=True)

median = df['Wait_Time (/Hrs)'].median()
median = df['Wait_Time (/Hrs)'].fillna(median, inplace=True)

median = df['Admission_Time (/Hrs)'].median()
median = df['Admission_Time (/Hrs)'].fillna(median, inplace=True)

median = df['Emergency_Visits'].median()
median = df['Emergency_Visits'].fillna(median, inplace=True)

median = df['Readmission (%)'].median()
median = df['Readmission (%)'].fillna(median, inplace=True)

median = df['Sepsis (/1000)'].median()
median = df['Sepsis (/1000)'].fillna(median, inplace=True)
```
